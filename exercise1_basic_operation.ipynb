## 7. 图片读写及显示

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('pic/bear500x333.jpg')
```

```python
cv.imwrite('test/bear.jpg', img)
```

```python
cv.imshow('bear', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img)
plt.show()
```

```python
# RGB BGR
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```

```python
img3 = cv.imread('pic/bear500x333.jpg', 0)
cv.imshow('bear', img3)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img3, cmap='gray')
plt.show()
```

```python
img3.ndim
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
apple = cv.imread('pic/apple.jpg')
show(apple)
```

```python
gray = cv.imread('pic/apple.jpg', 0)
show(gray)
```

## 9. 简单绘图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(2, 20)
y = 2 * x + np.random.randint(5, 20, 18)
```

```python
plt.plot(x, y, '+-', color='r')
plt.show()
```

```python
x = np.linspace(0, 1, 100)
y1 = np.power(x, 0.5)
y2 = x
y3 = np.power(x, 1.5)
```

```python
plt.plot(x, y1, label='0.5')
plt.plot(x, y2, label='1.0')
plt.plot(x, y3, label='1.5')
plt.legend()
plt.xlabel('r')
plt.ylabel('s')
plt.grid()
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.show()
```

```python
a = np.random.randint(0, 101, 1000)
```

```python
plt.hist(a, rwidth=0.9, color='g')
plt.show()
```

```python
bins = np.arange(-0.5, 101, 1)
bins
```

```python
plt.hist(a, bins)
plt.show()
```
